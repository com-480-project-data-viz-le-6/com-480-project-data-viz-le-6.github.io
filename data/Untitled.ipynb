```python
import pandas as pd
```

```python
df = pd.read_csv('wcm.csv') 
df
```

```python
df2020 = df[df["season"] == 2020]
df2020
```

```python
df_rank = pd.read_csv('points.csv')#, index_col=0) 
```

```python
df2020["ath_rank"]
```

```python
df_rank["ath_rank"]
```

```python
import numpy as np
df2020["ath_rank"]=df2020["ath_rank"].apply(lambda x: np.where(x.isdigit(),x,'0')).astype(int)
```

```python
df_total = pd.merge(df2020, df_rank, on="ath_rank").sort_values(by=['date'])
df_total
```

```python
df_total["cum_points"] = df_total.groupby(['ath_name'])["points"].cumsum()
```

```python
df_total[df_total["ath_name"] == "Alexis Pinturault"]
```

```python
df_total[["date", "ath_name", "country", "cum_points"]].to_csv("wcm2020.csv", index=False,)
```
